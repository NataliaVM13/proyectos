## **<font color="#000000"> Qu es SQL?</font>**
</td>
</tr>
</table>

**SQL** (Structured Query Language) es un lenguaje de programacin diseado para gestionar y manipular bases de datos relacionales.

**SQL** es fundamental en la gestin de datos, ya que permite a las organizaciones almacenar, recuperar y manipular informacin de manera efectiva y segura.

Es ampliamente utilizado en sistemas de gestin de bases de datos como MySQL, PostgreSQL, Microsoft SQL Server y Oracle.

<tr>
<td bgcolor='#B00A0F'>

### **<font color="#FFFFFF"> Caractersticas </font>**
</td>
</tr>
</table>

>* **Declarativo**:
  * SQL es un lenguaje declarativo, lo que significa que los usuarios especifican qu datos desean obtener o manipular, sin necesidad de definir cmo se deben realizar estas operaciones.
>* **Estndar**:
  * SQL es un estndar reconocido por ANSI (American National Standards Institute) e ISO (International Organization for Standardization), lo que permite su uso en diversas plataformas de bases de datos.
>* **Interoperabilidad**:
  * Puede ser utilizado en diferentes sistemas de gestin de bases de datos, como MySQL, PostgreSQL, Microsoft SQL Server, Oracle, entre otros.
